```python
# Define the Google Drive file URL
url = "https://drive.google.com/file/d/1vlhZ5c7abUKF8xXERIw6m9Te8fW7ohw3/view?usp=sharing"

# Extract the file ID from the URL (second last part of the path)
file_id = url.split("/")[-2]
print(file_id)  
```

```python
import gdown
```

```python
# Create the direct download link prefix for Google Drive
prefix = 'https://drive.google.com/uc?export=download&id='

# Use gdown to download the file by combining prefix + file_id
# The file will be saved locally as "kidney-CT-Scan-data.zip"
gdown.download(prefix + file_id, "kidney-CT-Scan-data.zip")
```
